```python
pivot = pd.pivot_table(index='TypeName', columns='Screen', data=df, values='Price_euros', aggfunc='count')
pivot
```

```python
pivot.plot(kind='bar', figsize=(12,6))
plt.title('count of price euros by typename and screen')
plt.xlabel('Typename')
plt.ylabel('Screen')
plt.xticks(rotation=45)
plt.legend(title='Screen')
plt.show()
```

```python
sb.heatmap(df.corr(numeric_only=True), annot=True, fmt='.1f', linewidths=2)
plt.xticks(rotation=45)
plt.show()
```

```python
sb.countplot(df, x='GPU_company', color='skyblue')
plt.title('count of GPU_company')
plt.show()
```

```python
sb.scatterplot(df, x='Weight', y='Price_euros')
plt.title('price euros and weight')
plt.show()
```

```python
df['Company'].value_counts().sort_values(ascending=True).plot(kind='barh', color='dodgerblue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title("Name of company and How many devices")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
df['TypeName'].value_counts().sort_values(ascending=True).plot(kind='barh', color='dodgerblue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title('count of type name')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
df['GPU_model'].value_counts()[:20].sort_values(ascending=True).plot(kind='barh', color='dodgerblue')
plt.grid(axis='y', linestyle='-.', alpha=0.7)
plt.title("count of GPU_model")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
